```python
#Import CSV File
df = pd.read_csv("/content/crowdfundingdataset.csv", encoding="latin1")
```

```python
#missing values
print(df.isnull().sum())
```

```python
#duplicate values
print(df.duplicated().sum())
```

```python
#separates numerical columns
num_cols = df.select_dtypes(include=['number']).columns
print(num_cols)
```

```python
#separates category columns
cat_cols = df.select_dtypes(include=['object']).columns.drop('CampaignID')
print(cat_cols)
```

```python
#checking unique values for categorical columns
df[cat_cols].nunique()
```

```python
#label encoding for categroical columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
print(df.head())
```

```python
#feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
X = df.drop(columns=['CampaignID', 'IsSuccessful'])
y=df["IsSuccessful"]
selector = SelectKBest(score_func=f_classif, k=6)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Best Features:", selected_features.tolist())
```

```python
#Split the Dataset: Divide the dataset into training and testing sets.
#Use 80% of the data for training and 20% for testing.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected,y, test_size=0.2, random_state=42)
```

```python
#Total data 100000 splited into 80% -80,000 and 20% -20,000
print("X_train",X_train.shape)
print("X_test",X_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
# Feature Scaling: Standardize columns using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Task 2 – Classification Model Building - RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix
rf_model = RandomForestClassifier(n_estimators=200,random_state=42)
```

```python
# Task 2 – Classification Model Building - RandomForestClassifier
rf_model.fit(X_train, y_train)
```

```python
# Task 2 – Classification Model Building - RandomForestClassifier
rf_pred = rf_model.predict(X_test)
```

```python
# Task 3 – Model Evaluation - RandomForestClassifier
rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_acc)
rf_cm = confusion_matrix(y_test, rf_pred)
print("\nConfusion Matrix for Randomforest:",rf_cm)
rf_precision = precision_score(y_test, rf_pred)
print("\nPrecision for Randomforest:",rf_precision)
rf_recall = recall_score(y_test, rf_pred)
print("\nRecall for Randomforest:",rf_recall)
rf_f1 = f1_score(y_test, rf_pred)
print("\nF1 Score for Randomforest:",rf_f1)
```

```python
# KNN
# Task 2 – Classification Model Building - KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7,weights='distance')
```

```python
# Task 2 – Classification Model Building - KNN
knn.fit(X_train_scaled, y_train)
```

```python
# Task 2 – Classification Model Building - KNN
knn_pred = knn.predict(X_test_scaled)
```

```python
# Task 3 – Model Evaluation - KNN
knn_acc = accuracy_score(y_test, knn_pred)
print("Accuracy for KNN:", knn_acc)
knn_cm = confusion_matrix(y_test, knn_pred)
print("\nConfusion Matrix for KNN:",knn_cm)
knn_precision = precision_score(y_test, knn_pred)
print("\nPrecision for KNN:",knn_precision)
knn_recall = recall_score(y_test, knn_pred)
print("\nRecall for KNN:",knn_recall)
knn_f1 = f1_score(y_test, knn_pred)
print("\nF1 Score for KNN:",knn_f1)
```

```python
#decision tree
# Task 2 – Classification Model Building - DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=42)
```

```python
# Task 2 – Classification Model Building - DecisionTreeClassifier
dt.fit(X_train, y_train)
```

```python
# Task 2 – Classification Model Building - DecisionTreeClassifier
dt_pred = dt.predict(X_test)
```

```python
# Task 3 – Model Evaluation -  DecisionTreeClassifier
dt_acc = accuracy_score(y_test, dt_pred)
print("Accuracy for decision tree:", dt_acc)
dt_cm = confusion_matrix(y_test, dt_pred)
print("\nConfusion Matrix for Decision tree:",dt_cm)
dt_precision = precision_score(y_test, dt_pred)
print("\nPrecision for Decision tree:",dt_precision)
dt_recall = recall_score(y_test, dt_pred)
print("\nRecall for Decision tree:",dt_recall)
dt_f1 = f1_score(y_test, dt_pred)
print("\nF1 Score for Decision tree:",dt_f1)
```

```python
#Naive Bayes Classifier
# Task 2 – Classification Model Building - Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
# Task 2 – Classification Model Building - Naive Bayes Classifier
nb.fit(X_train, y_train)
```

```python
# Task 2 – Classification Model Building - Naive Bayes Classifier
nb_pred = nb.predict(X_test)
```

```python
# Task 3 – Model Evaluation -  Naive Bayes Classifier
nb_acc = accuracy_score(y_test, nb_pred)
print("Accuracy for Naive bayes:", nb_acc)
nb_cm = confusion_matrix(y_test, nb_pred)
print("\nConfusion Matrix for Naive bayes:",nb_cm)
nb_precision = precision_score(y_test, nb_pred)
print("\nPrecision for Naive bayes:",nb_precision)
nb_recall = recall_score(y_test, nb_pred)
print("\nRecall for Naive bayes:",nb_recall)
nb_f1 = f1_score(y_test, nb_pred)
print("\nF1 Score for Naive bayes:",nb_f1)
```

```python
# Task 2 – Classification Model Building - Support Vector Machine
from sklearn.svm import SVC
svm = SVC(kernel='rbf',C=10,gamma='scale')
```

```python
# Task 2 – Classification Model Building - Support Vector Machine
svm.fit(X_train_scaled, y_train)
```

```python
# Task 2 – Classification Model Building - Support Vector Machine
svm_pred = svm.predict(X_test_scaled)
```

```python
# Task 3 – Model Evaluation - Support Vector Machine
svm_acc = accuracy_score(y_test, svm_pred)
print("Accuracy for SVM:", svm_acc)
svm_cm = confusion_matrix(y_test, svm_pred)
print("\nConfusion Matrix for SVM:",svm_cm)
svm_precision = precision_score(y_test, svm_pred)
print("\nPrecision for SVM:",svm_precision)
svm_recall = recall_score(y_test, svm_pred)
print("\nRecall for SVM:",svm_recall)
svm_f1 = f1_score(y_test, svm_pred)
print("\nF1 Score for SVM:",svm_f1)
```

```python
# Task 4 – Cross-Validation & Hyperparameter Tuning
from sklearn.model_selection import cross_val_score
models = {
    "Random Forest": rf_model,
    "Decision Tree": dt,
    "Naive Bayes": nb
}

print("Cross Validation Results\n")
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(name)
    print("Scores:", scores)
    print("Average Accuracy:", scores.mean())
```

Random Forest

Strengths:

* High classification accuracy.
* Reduces overfitting by combining multiple decision trees.
* Works well with large datasets.

Weaknesses:

* Requires more memory and computation time than a single decision tree.
* Less interpretable than simpler models.

Possible Improvements:

* Tune hyperparameters such as the number of trees (n_estimators) and maximum tree depth (max_depth).
* Perform additional feature engineering to improve performance.


K-Nearest Neighbors (KNN)

Strengths:

Simple to understand and implement.
Performs well when the data is properly scaled.

Weaknesses:

Prediction is slow for large datasets.
Sensitive to the choice of the number of neighbors (k) and feature scaling.

Possible Improvements:

Optimize the value of k using GridSearchCV.
Remove irrelevant features to improve prediction speed.

Decision Tree

Strengths:

Easy to understand and visualize.
Handles both numerical and categorical data.

Weaknesses:

Can easily overfit the training data.
Performance may decrease on unseen data if the tree becomes too deep.

Possible Improvements:

Limit the maximum depth of the tree.
Tune parameters such as max_depth and min_samples_split.

Naive Bayes

Strengths:

Fast to train and predict.
Works well on large datasets.
Requires very little computational power.

Weaknesses:

Assumes that all features are independent, which may not be true in real datasets.
May produce lower accuracy than more complex models.

Possible Improvements:

Apply better feature selection.
Tune the var_smoothing parameter using GridSearchCV.


Support Vector Machine (SVM)

Strengths:

Produces high accuracy for many classification problems.
Performs well with high-dimensional data.
Effective when the data is properly scaled.

Weaknesses:

Training is slow for large datasets.
Requires feature scaling before training.
Hyperparameter tuning can be computationally expensive.

Possible Improvements:

Optimize the C, kernel, and gamma parameters.
Use RandomizedSearchCV for faster hyperparameter tuning on large datasets.
Overall Conclusion



Among all the classifiers, SVM Model with the highest testing accuracy and F1-score was selected as the best-performing model for predicting crowdfunding campaign success.
